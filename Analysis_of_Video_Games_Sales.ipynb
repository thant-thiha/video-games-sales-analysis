## Data Preprocessing

```python
#Check Data Types
df.dtypes
```

A mismatch between data types in a given data dictionary and the actual dataset occurs when the field descriptions or expected data types (e.g., integer, float, string) listed in the data dictionary do not align with the true types present in the working data file. Aligning the data dictionary and the actual dataset types is essential for reliable, reproducible and accurate data analytics workflows. It prevents confusion, facilitates error-free processing and helps everyone working with the dataset to understand, use, and communicate insights effectively (Stirk, 2009). That's why we convert `Year_of_Release` as Integer, `Critic_Score` as Integer, `Critic_Count` as Integer, `User_Count` as Integer and `User_Score` as float.

```python
# Convert Year_of_Release from float to integer
df['Year_of_Release'] = df['Year_of_Release'].astype('Int64')

# Convert Critic_Score from float to integer (preserving NaNs)
df['Critic_Score'] = df['Critic_Score'].round(0).astype('Int64')

# Convert Critic_Count from float to integer (preserving NaNs)
df['Critic_Count'] = df['Critic_Count'].round(0).astype('Int64')

# Convert User_Count from float to integer (preserving NaNs)
df['User_Count'] = df['User_Count'].round(0).astype('Int64')

# Convert User_Score from object to float (handling 'tbd' or invalid entries)
def to_float(score):
    try:
        return float(score)
    except:
        return np.nan

df['User_Score'] = df['User_Score'].apply(to_float)
```

```python
df.dtypes
```

Changing the data types of the above columns ensures each field matches the definition in the data dictionary, integers for counts and years, and float for user scores. This step maintains data integrity, prevents errors during analysis, and ensures calculations and summaries are reliable for further business insights (Matillion, 2024).

```python
# Check missing values
missing_data = df.isnull().sum()
missing_percentage = (missing_data / len(df)) * 100
missing_summary = pd.DataFrame({
    'Missing Count': missing_data,
    'Missing Percentage': missing_percentage
})
print(missing_summary)
```

Several data quality issues were identified that require systematic handling. Key missing values include `Name` (2) and `Genre` (2). Additionally, there are 269 games without `Year_of_Release`, 8,580+ games missing `Critic_Score` and counts, and over 6,000 games missing `User_Score`. Missing `Publisher` information was found in 54 games, while `Developer` names and user age `Rating` were missing for over 6,600 entries.

```python
# Replace missing values with 'Unknown' instead of dropping
df['Name'] = df['Name'].fillna('Unknown')
df['Genre'] = df['Genre'].fillna('Unknown')
df['Year_of_Release'] = df['Year_of_Release'].fillna(pd.NA).astype('Int64')
df['Publisher'] = df['Publisher'].fillna('Unknown')
df['Developer'] = df['Developer'].fillna('Unknown')
df['Rating'] = df['Rating'].fillna('Not Rated')
```

```python
df[['Critic_Score', 'Critic_Count', 'User_Score', 'User_Count']].replace([None, '', 'tbd'], np.nan)
```

Replacing missing `Name`, `Genre`, `Year_of_Release`, `Publisher`, and `Developer` values with 'Unknown', and `Rating` with 'Not Rated' preserves all records for further analysis while clearly indicating missing information. This approach also avoids data loss, especially when the dataset is large and dropping rows or replacing with some valeus like means or medians may discard potentially valuable data. It also allows later filtering or special handling of only these unknown entries.

Replacing values like None, empty strings '', or 'tbd' with NaN in `Critic_Score`, `Critic_Count`, `User_Score`, and `User_Count` standardizes missing data representation and keep the data type as Integer and Float. This ensures these placeholders are still recognized as missing by pandas and analytical tool and also avoids errors caused by invalid non-numeric entries and supports consistent, unbiased data analysis (Little and Rubin, 2023).

```python
# Check missing values again
missing_data = df.isnull().sum()
missing_percentage = (missing_data / len(df)) * 100
missing_summary = pd.DataFrame({
    'Missing Count': missing_data,
    'Missing Percentage': missing_percentage
})
print(missing_summary)
```

## 2. Is there a correlation between the “eu_sales” and “jp_sales” for the years 2010-2014?

```python
# Filter for years 2010 to 2014
df_years = df[(df['Year_of_Release'] >= 2010) & (df['Year_of_Release'] <= 2014)]

# Calculate correlation
corr_value = df_years['EU_Sales'].corr(df_years['JP_Sales'])
print(f'Correlation between EU_Sales and JP_Sales (2010-2014): {corr_value:.2f}')
len(df_years)
```

```python
plt.figure(figsize=(7, 5))
plt.scatter(df_years['EU_Sales'], df_years['JP_Sales'], alpha=0.4)

# Add trendline
z = np.polyfit(df_years['EU_Sales'], df_years['JP_Sales'], 1)
p = np.poly1d(z)
plt.plot(df_years['EU_Sales'], p(df_years['EU_Sales']), color='red', linewidth=2, label='Minimal Relationship Line')

plt.xlabel('EU Sales (millions)')
plt.ylabel('Japan Sales (millions)')

# correlation for annotation
plt.title('Weak Correlation Between EU and Japan Video Game Sales (2010–2014)', pad = 12)
plt.grid(True, linestyle='--', alpha=0.6)

# Add the correlation value inside the plot (top right corner)
plt.annotate(f'Correlation: {corr_value:.2f}', 
             xy=(0.97, 0.97), 
             xycoords='axes fraction', 
             ha='right', va='top', 
             fontsize=10, 
             bbox=dict(boxstyle='round,pad=0.2', fc='white', ec='gray', lw=0.5))

plt.legend()
plt.show()
```

For Question 2, a focused subset of the dataset was analyzed by filtering only records where the year of release fell only between 2010 and 2014. This was implemented in Python using boolean indexing, ensuring all included sales data related specifically to that time frame. This targeted filtering enabled the analysis to reflect recent market trends and minimized confounding from legacy data.

We also noticed that some games are region-specific and may not be released or marketed in both Europe and Japan, which naturally results in zero sales for those regions. These zero values do not necessarily indicate missing or invalid data—they reflect genuine circumstances where the game wasn’t available or popular in those markets. Since we lack detailed background information about each game's regional release strategy, excluding these zeros could unintentionally remove valid data points and bias the correlation results. To ensure a fair and representative analysis, it is better to include the zero values, as they provide a complete picture of overall sales behavior across different markets.

The Pearson correlation coefficient was then calculated between `EU_Sales` and `JP_Sales` for these years using pandas’ `.corr()` method which is an appropriate choice to quantify the linear association between these two regional markets. A trendline was computed using `numpy.polyfit` for least-squares regression, allowing statistical relationships to be visualized clearly additionally.

A scatter plot was selected as it is ideally suited for exploring relationships between two continuous variables. Each point represents a unique video game's sales in the EU and Japan, and the dispersion visually communicates the strength and direction of correlation. This format enables the audience to intuitively assess the presence of any patterns (clusters, outliers, spread) (Ampler, 2025).

**Figure Size**: (7, 5) was chosen to optimize label space and ensure dense datapoints remain readable.

**Point Color & Transparency**: A light blue (alpha=0.4) was used for the markers to allow overlap visualization while keeping the chart visually soft and accessible.

**Trend Line**: The regression line was also added in addition to the data points and colored bright red and thickened to highlight the minimal positive relationship visually, making it easily distinguishable.

**Gridlines**: Dashed (linestyle='--') and semi-transparent (alpha=0.6) gridlines were added to support accurate reading of values and align with modern data visualization standards.

**Axis Labels**: 'EU Sales (millions)' and 'Japan Sales (millions)' were chosen to specify units for both axes, directly linking to the dataset and reducing ambiguity for viewers.

## 5. What are the 3 most common “Platform” in the dataset?

```python
# Find the 3 most common platforms by count and group the rest as 'Others'
top_platforms = df['Platform'].value_counts().head(3)
other_count = df['Platform'].value_counts().iloc[3:].sum()

# Combine the top platforms and 'Others' category
top_platforms_with_others = pd.concat([top_platforms, pd.Series({'Others': other_count})])

vals = top_platforms_with_others.values
labs = top_platforms_with_others.index
```

```python
explode = [0.08, 0.08, 0.08, 0]  # light explode for top 3 only

def autopct_with_counts(pct):
    total = vals.sum()
    count = pct*total/100.0
    # Show percentage to 1 decimal, count with commas
    return f'{pct:.1f}%\n({count:,.0f})'

plt.figure(figsize=(6, 5))
wedges, texts, autotexts = plt.pie(
    vals,
    labels=labs,
    autopct=autopct_with_counts,
    startangle=120,
    colors=['#2978B5', '#3DD597', '#FF968A', '#E0E0E0'],  
    pctdistance=0.75,
    labeldistance=1.05,
    wedgeprops=dict(edgecolor='white', linewidth=3),
    explode=explode
)
centre_circle = plt.Circle((0, 0), 0.60, fc='white') 
plt.gca().add_artist(centre_circle)
plt.setp(autotexts, size=10, color='#222')
plt.setp(texts, size=12)
# Main title as "supertitle" (at the top)
plt.suptitle('Platform Leaders: PS2, DS, and PS3 top the catalog', fontsize=15, y=0.96)

# Subtitle below title, adjusted with 'y' value for placement just above donut
plt.title('These three platforms together account for over one-third of all titles', fontsize=10, color='dimgray', loc='center', pad=10)

plt.axis('equal')
plt.tight_layout()
plt.show()
```

To visually highlight platform trends in the dataset, the three most common platforms were identified by calculating value counts for the `Platform` column and selecting the top three entries. All other platforms were combined into an "Others" category by summing the remaining counts. Including "Others" lets the audience see how much of the "pie" is left after the top categories, supporting honest reporting and effective visual communication rather than just top 3 especially if we are going to visualize it in pie.

This aggregation, accomplished using pandas methods (`value_counts`, `head`, and `iloc`), condenses a potentially overwhelming number of unique platforms into a more digestible summary. The final categorical distribution was consolidated into a new Series (top_platforms_with_others), providing clear focus for later analysis.

A donut-style pie chart was chosen to represent the relative proportions of platform counts, as this format excels at communicating part-to-whole relationships. Pie charts are immediately intuitive for most audiences and, by "exploding" (offsetting) the top three wedges, added emphasis is placed on the leading platforms. The donut variant improves readability, enables easy comparison, and creates space for bold internal labeling (Sage Journals, 2024).

The "Others" category is used in a pie chart to group all the less common categories that don't make the top ranking (platforms), which improves clarity and accuracy. Without "Others," the chart would misleadingly suggest that only the top three categories make up 100% of the total when, in fact, a large portion belongs to less frequent categories. 

**Explode**: A slight offset (explode=[0.08, 0.08, 0.08, 0]) for the top three wedges visually lifts them out, drawing attention to the platforms of primary interest without overshadowing the "Others" group.

**Color**: Distinct and contrasting colors (#2978B5, #3DD597, #FF968A, and a subdued gray for "Others") assign unique identities to each wedge, helping the viewer distinguish between groups while maintaining visual harmony.

## 6. Statistically Relevant Question
The analytical question posed was:
> **How have video game sales evolved over the years, and are there statistically significant upward or downward trends in each NA, EU, JP and Other region sales?**

Time series trend analysis was chosen because it reveals how video game sales have changed year-over-year—insight that can not be gained by category or correlation analysis that we did above alone. It also builds on previous questions which focused on “top” publishers, correlations between markets, genre distributions, and period comparisons by examining long-term performance. This answers new questions such as "Are industry sales rising or falling?" "Are some regions more resilient or volatile?". Regression results give not only a visual but a quantitative basis for claims, showing that observed declines or growth (slope) are reliable (p-value) and how tightly years predict sales (r-value) (Li et al., 2007).

By addressing trends and their statistical significance, the analysis deeply enhances insight into the video game market, equipping decision-makers to anticipate future shifts, contextualize historic events, and strategize using both trends and regional detail.

To answer this, annual sales data for each region except global sales were aggregated using Python’s pandas library, enabling systematic grouping and summing of sales by year and region. Leveraging the power of the `scipy.stats.linregress` function, linear regression was conducted for each sales region. This step produced the slope (trend direction and rate), p-value (significance), and r-value (correlation strength) for each region’s yearly sales, providing a robust statistical foundation for the analysis as below.

**Note**: During the data preparation phase, it was observed that 269 entries in the `Year_of_Release` column contained missing values. These rows were excluded from the time series trend analysis, meaning the associated sales figures could not be allocated to specific years. As a result, the annual sales totals and thus the statistical conclusions about market trends—represent only the subset of records with valid year information.

This exclusion can lead to slight underestimation of yearly sales figures, particularly if missing years are not randomly distributed. For example, if the missing data are concentrated in certain time periods or involve high-selling titles, some patterns or peaks could be muted in the final results. However, since the proportion of missing values is relatively low compared to the overall dataset size, the analysis still robustly captures long-term sales movements and major trends.

Additionally, only complete data up to 2016 are available in the dataset. For this reason, the visualization and statistical trend analyses are limited to years with full data coverage, and sales beyond 2016 are not displayed or interpreted to ensures that all conclusions are based on consistently reported and reliable data.

```python
# Keep only years from 2016 backwards (2016-1980 inclusive)
valid_years = list(range(1980, 2017))
df = df[df['Year_of_Release'].isin(valid_years)]

# Aggregate sales by year
sales_by_year = df.groupby('Year_of_Release')[['NA_Sales', 'EU_Sales', 'JP_Sales', 'Other_Sales']].sum()
```

```python
from scipy.stats import linregress
# Linear regression slope for each sales region
trends = {}
for col in sales_by_year.columns:
    x = sales_by_year.index.values.astype(float)
    y = sales_by_year[col].values
    slope, intercept, r_value, p_value, std_err = linregress(x, y)
    trends[col] = {'slope': slope, 'p_value': p_value, 'r_value': r_value}
```

```python
# Put the linear regression into table
trends = pd.DataFrame(trends).T
trends.reset_index(inplace=True)
trends.rename(columns={'index': 'Region'}, inplace=True)
print(trends)
```

```python
ax = sales_by_year.plot(
    kind='line',
    marker='o',
    linewidth=2,
    figsize=(8, 6),
    color=['#1f77b4', '#ff7f0e', '#2ca02c', '#d62728', '#9467bd']
)
plt.title('Video Game Sales Trends by Region (1980–2016)', fontsize=15, pad=16)
plt.xlabel('Year of Release', fontsize=12)
plt.ylabel('Total Sales (millions)', fontsize=12)
plt.yticks(np.arange(0, plt.ylim()[1]+25, 25))
plt.legend(['NA Sales', 'EU Sales', 'JP Sales', 'Other Sales'], loc='upper right', fontsize=9)
plt.grid(axis='y', linestyle='--', alpha=0.6)

# Add concise overall trend summary to upper left
plt.text(
    1980,                 # x coordinate 
    plt.ylim()[1] * 0.70, # y coordinate 
    "Sales surged in NA and EU,\npeaked in the 2000s, then declined.\nJP and Other regions had steady, slower growth.",
    fontsize=11,
    va='top',
    ha='left',
    bbox=dict(facecolor='white', alpha=0.5, edgecolor='gray', boxstyle='round,pad=0.2')
)

plt.tight_layout()
plt.show()
```

Line charts are well-suited for identifying rises, falls, cycles, and peaks—patterns that are not visible in bar or pie charts. Displaying all regions on the same axes places sales trends in direct context, letting the viewer assess which regions grew or declined together or diverged (Chen et al., 2007).

**Color Palette**: Each region is assigned a distinct, colorblind-friendly color, making trends easy to distinguish and compare. Consistent with best practices, these colors are chosen for clarity and accessibility.

**Main Title**: Clearly states the plot’s focus (regional sales trends and period), orienting viewers immediately. Prominent size and spacing separate the title from the plot, following recommendations for effective, professional visual hierarchy.

**Axes Labels**: Precisely define each plotted dimension, making the graph self-explanatory. Font size balances readability with unobtrusiveness.

**Y-axis Grid**: Intervals of 25 million units give regular, logical reference points for sales volumes, easing value estimation across the plot. Also dashed, semi-transparent lines for non-intrusive, background support, helping the eye track across to the axis for quick comparison.

**Legend Location**: Placed away from data-dense areas (upper right) to avoid overlap and confusion. Font Size is Slightly smaller to keep attention on main plot, but large enough for quick reference.

**Explanatory Text Box**: Positioned in the upper left, this high-level narrative guides the viewer’s interpretation by summarizing the main trends and keeping the analysis visually anchored in context. Soft background and rounded edges ensure it stands out while blending naturally into the plot, supporting but not overpowering the figure.

**Layout Optimization**: Automatically adjusts plot elements for the best fit within the figure, preventing overlap and ensuring all content remains accessible (Chen et al., 2007).
